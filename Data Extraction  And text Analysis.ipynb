```python
import requests
from bs4 import BeautifulSoup

reports = []
for url in df.SECFNAME:
    r = requests.get(url)
    data = r.text
    soup = BeautifulSoup(data, "html.parser")
    reports.append(soup.get_text())

print(f'Total {len(reports)} reports saved')
```

```python
for report in reports:

    ps= PorterStemmer()
    wordnet = WordNetLemmatizer()
    sentences = nltk.sent_tokenize(report)
    corpus = []
    for i in range(len(sentences)):
        review=re.sub('[^a-zA-Z]',"",sentences[i])
        review=review.lower()
        review=review.split()
        review=[ps.stem(word) for word in review if not  word in set(stopwords.words('english'))]
        review=' '.join(review)
        corpus.append(review)
```

```python
#tokenization and removing stop_words

def tokenize(text):
    text = re.sub(r'[^A-Za-z]',' ',text.upper())
    tokenized_words = nltk.word_tokenize(text)
    return tokenized_words

def remove_stopwords(words, stop_words):
    return [x for x in words if x not in stop_words]







#counting no. of wordsin specific

def countfunc(s_dict,words):
    count=0
    for word in words:
        if word in s_dict:
            count=count+1
    return count


#polarity
#from textblob import TextBlob

#def get_polarity(words):
 #   return TextBlob(words).sentiment.polarity

#subjectivity
#def get_subjectivity(words):
 #   return TextBlob(words).sentiment.subjectivity


def polarity(positive_score,negative_score):
    polarity = ((positive_score-negative_score)/((positive_score+negative_score)+ 0.000001))
    return polarity

def subjectivity(positive_score, negative_score,num_words):
                return (positive_score+negative_score)/(num_words+0.000001)


def sentiment(score):
    if(score < -0.5):
        return 'Most Negative'
    elif(score >= -0.5 and score < 0):
        return 'Negative'
    elif(score == 0):
        return 'Neutral'
    elif(score > 0 and score < 0.5):
        return 'Positive'
    else:
        return 'Very Positive'
    

#for complexword determination    
def syllabus(word):
    if (len(word) > 2 and (word[-2:]=='ed' or word[-2:]=='es')):
        return True
    
    count=0
    vowels = ['a','e','i','o','u']
    for i in word:
        if i.lower() in vowels:
            count=count+1
    if (count > 2):
        return True
    else:
        return False
            
def fog_index_cal(average_sentence_length, percentage_complexwords):
    return 0.4*(average_sentence_length + percentage_complexwords)
```
